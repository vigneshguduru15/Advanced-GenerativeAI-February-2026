Problem 1 — Smart Parking Lot Management

```python
def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")


capacity = 50
logs = ["IN","IN","IN","OUT","IN","IN","OUT"]

parking_system(capacity, logs)
```

Problem 2 — Online Food Delivery Time Estimator

```python
def estimate_delivery(distance, traffic, weather):
    base_time = distance * 5

    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 10

    if weather.lower() == "rainy":
        base_time += 10

    print("Estimated Delivery Time:", base_time, "minutes")


estimate_delivery(8, "High", "Rainy")
```

Problem 3 — Movie Theatre Seat Occupancy

```python
def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    percentage = (booked / total_seats) * 100

    print("Occupancy:", round(percentage), "%")

    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")


booked_seats = [1]*150
seat_occupancy(200, booked_seats)
```

Problem 4 — Cloud Server Load Classification

```python
def server_load(cpu_readings):
    avg = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif 50 <= avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


server_load([45,60,70,85,90])
```

Problem 5 — Smart Classroom Resource Monitor

```python
def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")


usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

Problem 6 — Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
    else:
        confirmed = capacity
        waitlist = registrations - capacity

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)

    if registrations > capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


event_registration(100, 105)
```
