Problem 1: Social Media – Post Engagement Analyzer


```python
def analyze_engagement(likes_list):
    total_likes = sum(likes_list)

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")
analyze_engagement([200, 300, 600])
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_analysis(rainfall, required):
    total = 0
    for r in rainfall:
        total += r

    avg = total / len(rainfall)

    if avg >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print(f"Average Rainfall: {avg}")
    print(f"Rainfall Status: {status}")

rainfall_analysis([70, 75, 72], 70)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

duplicate_accounts(["user1", "user2", "user1"])
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
def premium_crop_filter(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print(f"Premium Crops: {premium}")

premium_crop_filter([1500, 2500, 3200])
```

Problem 7: System Monitoring – Application Health Checker

```python
def app_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {errors}")
    print(f"System Status: {status}")

app_health(7)
```

Problem 9: E-Learning – Student Attendance Eligibility System

```python
def attendance_eligibility(attendance):
    total = len(attendance)
    present = sum(attendance)

    percentage = (present / total) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")

attendance_eligibility([1,1,1,0,1,1,1,1])
```

Problem 10: Smart Electricity Bill Analyzer


```python
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print(f"Total Bill: ₹{bill}")
    print(f"Usage Level: {usage}")

electricity_bill(250)
```
