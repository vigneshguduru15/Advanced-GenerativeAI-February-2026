Problem 1 — Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Recharge must be at least ₹50")
        elif amount not in valid_plans:
            print("Invalid plan. Valid plans:", valid_plans)
        else:
            print("Recharge successful for ₹", amount)
            break

validate_recharge()
```

Problem 2 — Inventory Reorder Alert System

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")

products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 20
}

check_inventory(products)
```

Problem 3 — Student Result Processing System

```python
def student_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        print("Average:", average, "Pass")
    else:
        print("Average:", average, "Fail")

student_result([60, 55, 70, 40])
```

Problem 4 — Cab Fare Estimator

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Peak hour? (yes/no): ")

    fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry != "yes":
        break
```

Problem 5 — Employee Attendance Eligibility

```python
def check_attendance(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = ["P","P","A","P","P","A","P","P"]
result = check_attendance(attendance)

print("Attendance Status:", result)
```

Problem 6 — Password Strength Checker

```python
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


password = input("Enter password: ")
print(check_password(password))
```
